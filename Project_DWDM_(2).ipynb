```python
### function หา % ของข้อมูลที่จะหายไปถ้า .dropna()
def loss_data(data):
    before = data.shape[0]
    after = data.dropna().shape[0]
    c = before - after
    print(f'size before drop = {before}')
    print(f'size after drop = {after}')
    print(f'we loss {c*100/before} % of data')
```

```python
loss_data(data_03) # จะเห็นได้ว่าหลังจากที่ .dropna() ข้อมูลหายไป 55.84 % ซึ่งเป็นจำนวนมาก ดังนั้นจึงเลือกที่จะแก้ค่า missing ด้วยวิธีอื่น
```
